# Ventas vs. numero de lanzamientos

```python
agrupado_df = df.groupby('Year_of_Release').agg({'Name': 'count', 'Global_Sales': 'sum'}).reset_index()
normalized_df = (agrupado_df - agrupado_df.mean()) / agrupado_df.std()
normalized_df['year'] = agrupado_df['Year_of_Release']

plt.figure(figsize=(10, 5))
sns.pointplot(x=normalized_df.year, y=normalized_df.Name, color='blue', label='Release Count')
sns.pointplot(x=normalized_df.year, y=normalized_df.Global_Sales, color='red', label='Global Sales')

plt.legend(handles=[mpatches.Patch(color='blue', label='Numero de lanzamientos'),
                    mpatches.Patch(color='red', label='Ventas Globales')],
           loc='upper left', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

Las formas siguen una buena correspondencia entre s√≠, no se observa ninguna discrepancia extra√±a, como altas ventas con pocas publicaciones o viceversa. Otro aspecto a considerar es que el n√∫mero de lanzamientos tiene una tendencia m√°s uniforme en comparaci√≥n con las ventas, que parecen ser m√°s dif√≠ciles de predecir con precisi√≥n. En resumen, las tendencias entre la cantidad de lanzamientos y las ventas globales est√°n bien alineadas, aunque las ventas muestran m√°s variabilidad en comparaci√≥n con la fluidez del n√∫mero de lanzamientos.

# Ventas totales por genero




```python
genre_sales = df.groupby('Genre').Global_Sales.sum()

ax = genre_sales.sort_values(ascending=False).plot(kind='bar', figsize=(13, 5))

total_sum = genre_sales.sum()
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f"{round(height)}\n{round(height / total_sum * 100)}%",
                (p.get_x() * 1.007, height * 0.75),
                color='black')

plt.xlabel('Genero')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por genero')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Los resultados num√©ricos coinciden con los an√°lisis que hemos realizado hasta el momento. Es evidente que el g√©nero de "acci√≥n" lidera en t√©rminos de ventas totales. Este g√©nero parece tener una conexi√≥n particularmente s√≥lida con los jugadores, ya que ha logrado acumular la mayor cantidad de ventas en comparaci√≥n con otros g√©neros.

# Ventas regionales de generos

```python
genero_geo_rankings = [
    df.groupby('Genre').sum().unstack().NA_Sales.sort_values(ascending=False).index,
    df.groupby('Genre').sum().unstack().EU_Sales.sort_values(ascending=False).index,
    df.groupby('Genre').sum().unstack().Other_Sales.sort_values(ascending=False).index,
    df.groupby('Genre').sum().unstack().JP_Sales.sort_values(ascending=False).index
]

genre_geo_rank_df = pd.DataFrame(columns=['genre', 'rank_type', 'rank'])

for region, rankings in enumerate(genre_geo_rankings):
    for rank, genre in enumerate(rankings):
        genre_geo_rank_df = genre_geo_rank_df.append({'genre': genre, 'rank_type': region, 'rank': rank}, ignore_index=True)


plt.figure(figsize=(10, 8))
sns.barplot(x='rank_type', y=12 - genre_geo_rank_df['rank'], hue='genre', data=genre_geo_rank_df)

plt.xlabel('Region')
plt.ylabel('Clasificaci√≥n de G√©nero')
plt.title('Clasificaci√≥n de G√©nero por Region')

x_labels = ['NorteAmerica', 'Europa', 'Resto del mundo', 'Japan']
plt.xticks(range(4), x_labels)
plt.gca().invert_yaxis()


plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Se nota que los gustos de los jugadores en Jap√≥n son muy diferentes de los de las otras tres regiones. En cuanto a las otras tres, son m√°s o menos similares; tal vez solo los g√©neros de plataformas y carreras cambian su posici√≥n en m√°s de una regi√≥n.

# Gr√°fico de dispersi√≥n entre Ventas Globales y Ventas Regionales

```python
ventas_globales = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
ventas_normalizadas = (df[ventas_globales] - df[ventas_globales].mean()) / (df[ventas_globales].max() - df[ventas_globales].min())

markers = ['+', '.', 'x', 'o']
regiones = ['Norteam√©rica', 'Uni√≥n Europea', 'Jap√≥n', 'Otras Regiones']

plt.figure(figsize=(10, 6))
for col, marker, region in zip(ventas_normalizadas.columns[:-1], markers, regiones):
    sns.regplot(x=ventas_normalizadas.Global_Sales, y=ventas_normalizadas[col], marker=marker, label=region)

plt.xlabel('Ventas Globales Normalizadas')
plt.ylabel('Ventas Regionales Normalizadas')
plt.title('Relaci√≥n entre Ventas Globales y Ventas Regionales')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


# Plataforma m√°s vendida cada a√±o

```python
temp = df.groupby(['Year_of_Release', 'Platform']).sum().reset_index().groupby('Year_of_Release')

plataforma_anual_ganadora_df = pd.DataFrame()

for a√±o, grupo in temp:
    a√±o_actual = temp.get_group(a√±o)
    ventas_maximas_este_a√±o = 0.0
    ganadora_actual = ""
    fila = {'a√±o': "", 'ganadora': "", 'ventas': ""}
    for √≠ndice, datos_plataforma in a√±o_actual.iterrows():
        if datos_plataforma.Global_Sales > ventas_maximas_este_a√±o:
            ventas_maximas_este_a√±o = datos_plataforma.Global_Sales
            ganadora_actual = datos_plataforma.Platform

    fila['a√±o'] = a√±o
    fila['ganadora'] = ganadora_actual
    fila['ventas'] = ventas_maximas_este_a√±o
    plataforma_anual_ganadora_df = plataforma_anual_ganadora_df.append(fila, ignore_index=True)

plt.figure(figsize=(13, 4))

grafico = sns.pointplot(x=plataforma_anual_ganadora_df.a√±o,
                        y=plataforma_anual_ganadora_df.ventas,
                        hue=plataforma_anual_ganadora_df.ganadora,
                        markers=["o", "s", "D", "P", "^", "v", "<", ">", "h", "*", "x"])

grafico.set_xticklabels(grafico.get_xticklabels(), rotation=90)

plt.xlabel('A√±o')
plt.ylabel('Ventas Globales')
plt.title('Plataforma Ganadora de Ventas Cada A√±o')
plt.legend(title='Plataforma', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
