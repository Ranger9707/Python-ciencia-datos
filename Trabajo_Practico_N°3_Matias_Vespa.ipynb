***APLICACIÓN DE PROBABILIDAD***

Se Realiza el lanzamiento de un dado EQUILIBRADO JUSTO, de manera reiterada con las mismas condiciones, y se anotan sus resultados= CARA O CRUZ.

Se le pide al usuario que defina la cantidad N de lanzamientos y se calcula su frecuencia relativa= 'PROBABILIDAD' de obtener cara o cruz. Se obtiene los siguientes resultados:

```python
import numpy as np
n=int(input('ingrese la cantidad de simulación o lanzamientos'))

#========= FORMA 1 DE SIMULAR LANZAMIENTO DE UN DADO============
dado=[]
cara=0
cruz=0
for i in range(n):
  dado.append(np.random.rand(1))
  dado1=np.array(dado)
  if dado1[i]<0.5:
    #print('salio cara')
    #print(dado[i])
    cara +=1
  else:
    #print('salio cruz')
    #print(dado[i])
    cruz +=1
```

```python
#  mostrar la probailidad frecuencista de el lanzamiento de un dado
prom_cara=cara/n
prom_cruz=cruz/n
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CARA es:', prom_cara)
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CRUZ es:', prom_cruz)
```

```python
#========== FORMA COMPACTA DE SIMULAR USANDO NUMPY===========
m=int(input('ingrese la cantidad de simulación o lanzamientos'))
lanzamientos = np.random.randint(0, 2, m)

# Calcular las probabilidades

probabilidad_cara = np.sum(lanzamientos ==0 ) / len(lanzamientos)
probabilidad_sello = np.sum(lanzamientos== 1) / len(lanzamientos)

print(f"Probabilidad de cara: {probabilidad_cara:.2f}")
print(f"Probabilidad de sello: {probabilidad_sello:.2f}")
```

```python
import matplotlib.pyplot as plt
x=np.array(range(2))
y=np.array([probabilidad_cara, probabilidad_sello])

plt.bar(x,y)
plt.xlabel('Opción')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma del lanzamiento de una moneda equilibrada')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos


# Solicitar al usuario la cantidad N de datos a cargar
N = 1000

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)
```

```python
#Deteccion de Outliers
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
columns_to_check = ["Age"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)
outliers_df = pd.concat(outliers_list)
df_cleaned1 = df.drop(outliers_df.index)
print("Número de outliers detectados:", len(outliers_df))
print(outliers)
```

PassengerId: variable categorica nominal, categoría en la que las etiquetas son simplemente nombres o identificadores sin ningún orden inherente.

Survived:categórica nominal, ya que representa diferentes categorías sin un orden numérico entre ellas.

Pclass: Categorica Discreta: es un tipo de dato que sigue el orden natural o una clasificacion entre si, esta variable representa la clase dentro del  titanic segun el estatus social.

Sex:categórica nominal, ya que representa diferentes categorías sin un orden numérico entre ellas.

Age: Categorica discreta, no utiliza valores decimales

Fare: Numerica continua: Numeros incontables o de indeterminados valores, en este caso Fare hace referencia al precio de los boletos, que varian entre si y son diferentes hasta con 4 decimales de precision.

Embarked: Categorica nominal: Datos que no tienen ninguna relacion o orden entre si, esta variable representa el puerto de embarque de los pasajeros, datos irrelevantes entre si.

Decidi eliminar las siguientes variables:

Name: El nombre de una persona es único y no proporciona información significativa para análisis estadísticos. Además, extraer información útil de los nombres puede ser complicado y requerir técnicas de procesamiento de lenguaje natural más avanzadas.

Cabin: La información de la cabina puede estaba incompleta y tenia muchos valores faltantes.

Ticket: Los números de boletos pueden ser únicos y no seguir un patrón discernible. A menos que tengas información específica sobre la estructura de los números de boletos y cómo se relacionan con otras variables, se considero irrelevante.

Parch y SibSp: Estas variables se relacionan con información sobre familiares a bordo, como padres/hijos y hermanos/cónyuges. Se considero irrelevante.
